```python
text = "LangChain is the framework for building context-aware reasoning applications"
text2 = "Python is the framework for building context-aware reasoning applications"
text3 = "java is the framework for building context-aware reasoning applications"
two_vectors = embeddings.embed_documents([text, text2,text3])
for vector in two_vectors:
    print(vector[:100])  # Show the first 100 characters of the vector

question="What is java?"
query_embed=embeddings.embed_query(question)
print()
print(query_embed)


```
